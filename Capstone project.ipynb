# Metrics
In this project I will use the evaluation metric of fbeta_score. It measures the effectiveness of retrieval with respect to a user who attaches beta times as  much importance to recall as precision.                      
 
- Fβ=(1+β2)precisionrecall/(β2precision)+recall 

The F-beta score can be interpreted as a weighted harmonic mean of the precision and recall, where an F-beta score reaches its best value at 1 and worst score at 0.
When beta=0.5 more emphasis is placed on precision. This is called  f 0.5 score. For my project, I will calculate 
fbeta_score following sklearn documentation as follows 

 ### sklearn.metrics.fbeta_score(y_true, y_pred, beta, labels=None, pos_label=1, average=’binary’, sample_weight=None) 


In my project I will be calculating the fbeta_score with beta=0.5.

# Methodology

## DATA PREPROCESSING

### Removing the target variable from the data set

Here 'quality' is our target variable.So, we drop quality.

```python
y= pd.Series(data['quality'])
X=data.drop('quality', axis=1)
display(X.head())
```

### Missing Value Treatment

Now we will find the number of null values for each feature:

```python
data.isnull().sum()
```

Here we have no NULL values in our data set .So, we need not remove any of the values.

### Outlier Treatment:

```python
l=0
cnt=0
tot_outliers = np.array([])
l2=np.array([])
for val in X.keys():
    Q1= np.percentile(X[val], q=25)
    Q3= np.percentile(X[val],q=75)
    IQR = Q3 - Q1
    lower_bound = Q1 - (IQR * 1.5)
    upper_bound = Q3 + (IQR * 1.5)
    print("outliers for:{}".format(val)) 
    outliers = X[((X[val] < lower_bound) | (X[val] > upper_bound))]
    #display(outliers)
    
    #for i in outliers.index.get_values.unique():
        #if counter(i)
        
    tot_outliers = np.append(outliers.index.get_values(),tot_outliers)
com_outliers = []
#print(Counter(tot_outliers))
x=dict(Counter(tot_outliers))
for i in x:
    if x[i]>1:
        com_outliers.append(int(i))
print(com_outliers)
    

    
```

### Removing Outliers 

```python
# Remove the outliers, if any were specified

good_data = X.drop(data.index[com_outliers]).reset_index(drop = True)
```

```python
display(good_data.shape)
```

### Removing Outliers from Target variable 

```python
y=y.drop(com_outliers)
```

# Number of instances for each class

```python
sns.countplot(y)
plt.show()
```

### converting the output variable suitable for classification 

```python
y = pd.Series([ 1 if x > 5 else 0 for x in y ])
```

```python
sns.countplot(y)
plt.xlabel("classified values of the feature 'quality'")
plt.ylabel("Frequency")
plt.show()
```

## Feature Scaling (using MinMax scaler) 

Feature scaling is a method used to standardize the range of independent variables or features of data. In data processing, it is also known as data normalization and is generally performed during the data preprocessing step.

```python
# Initialize a scaler, then apply it to the features
scaler = MinMaxScaler() # default=(0, 1)

good_data[:]= scaler.fit_transform(good_data[:])

# Show an example of a record with scaling applied
display(good_data.head(5))
```

```python
grap = good_data.hist(figsize=(20,20))

for ax in grap.flatten():
    ax.set_xlabel('Concentration of the Components')
    ax.set_ylabel('Frequency')
plt.show()
```

In the above plots, we can see that all the feature's are ranging from [0,1].This is because of performing MinMax scaling.We are rescaling the range of features to scale the range in [0, 1] or [−1, 1]. Selecting the target range depends on the nature of the data.

### Train-Test Split

```python
X_train , X_test , y_train , y_test = train_test_split(good_data, y , test_size = 0.25 , random_state = 145)

print("TRAINING DATA")
display(X_train)
display(y_train)
print("TESTING DATA")
display(X_test)
display(y_test)
```

## Implentation

### Bench mark model-Logistic Regression

```python
LR =LogisticRegression(random_state=1)
LR.fit(X_train, y_train)
LR_predictions = LR.predict(X_test)

accuracy= fbeta_score(y_test, LR_predictions,beta=0.5)
print(accuracy)
```

### Testing different Algorithms 

# SVC

```python
svm_model_linear = SVC().fit(X_train, y_train)
svm_predictions = svm_model_linear.predict(X_test)
  
accuracy = fbeta_score(y_test, svm_predictions,beta=0.5)
print(accuracy)
```

# DecisionTree

```python
from sklearn.tree import DecisionTreeClassifier
dec = DecisionTreeClassifier(random_state=1)
dec.fit(X_train,y_train)
dec_pred = dec.predict(X_test)

accuracy = fbeta_score(y_test, dec_pred,beta=0.5)
print(accuracy)
```

# AdaBoost

```python
from sklearn.ensemble import AdaBoostClassifier 
dec = AdaBoostClassifier (random_state=1)
dec.fit(X_train,y_train)
dec_pred = dec.predict(X_test)

accuracy = fbeta_score(y_test, dec_pred,beta=0.5)
print(accuracy)
```

# RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
dec = RandomForestClassifier (random_state=1)
dec.fit(X_train,y_train)
dec_pred = dec.predict(X_test)

accuracy = fbeta_score(y_test, dec_pred,beta=0.5)
print(accuracy)
```

## Refinement 

# Grid Search for RandomForest

```python
from sklearn.cross_validation import ShuffleSplit
from sklearn.metrics import make_scorer

param_grid = {'n_estimators':[10,50,100,150],'max_depth': [1,2,3,4,5,8]}

clf=RandomForestClassifier()

scoring=make_scorer(fbeta_score,beta=0.5)
grid_search = GridSearchCV(clf, param_grid,cv=5,scoring=scoring)
kfit=grid_search.fit(X_train, y_train)
model = kfit.best_estimator_
print(model)
final = model.predict(X_test)
```

```python
final_score=fbeta_score(y_test,final,beta=0.5)
print(final_score)
```
